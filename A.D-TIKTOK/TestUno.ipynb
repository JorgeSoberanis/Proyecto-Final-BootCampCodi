```python
import pandas as pd
import numpy as np
import altair as alt
from vega_datasets import data
```

# Aqui lo que se hace es leer el archivo .csv y se imprimen los primeros cinco c:

```python
dt = pd.read_csv("tiktok.csv", index_col= "id")
dt
```

### Quiero eliminar unas columnas para que el datasets quede mas facil de leer, vamos a ver que tal 

```python
data = dt.drop(['track_id', "artist_id","album_id", "duration", "danceability", "energy", "key", "speechiness", "acousticness", "instrumentalness", "liveness", "playlist_name", "playlist_id"], axis=1)
```

## Primer grafico de barras que muestre los artistas del dataframe con Altair c:

```python
alt.Chart(data).mark_bar().encode(
    x= alt.X("popularity",title = "Popularidad"),
    y= alt.Y("artist_name:O", title= "Artista"),
    color= alt.condition(
        alt.datum.popularity >= 70,
        alt.value("orange"),
        alt.value("steelblue")
    )

).properties(
    title = f"Popularidad de los artistas en TikTok. Diciembre 2021",
    width = 600,
)
```
